## 一次性环境准备：数据入库 SQLite

```python
import pandas as pd
import sqlite3

# 连接数据库
conn = sqlite3.connect('retail_sales.db')

# 读取四张Excel表，并写入sqlite
path = "/kaggle/input/datasets/tclaim/retail-business-sql-analysis-dataset/SQL_.xlsx"
sheet_names = ["销售订单表","商品信息表","客户信息表","门店信息表"]
for sheet_name in sheet_names:
    df = pd.read_excel(path, sheet_name)
    df.to_sql(sheet_name, conn, index=False, if_exists="replace")

conn.close()
print("原始业务表已成功写入SQLite，所有模块可直接独立运行")
```
