## 模块 1：全量业务核心指标一键统计
### 业务需求

基于原始表，一键统计零售业务全量核心基础指标，快速建立业务全貌认知

### 思路分析

1. 对 4 张原始表分别做基础聚合，用`UNION ALL`合并结果集，单段代码完成全量指标统计

2. 覆盖业务规模、数据周期、核心交易指标

### 核心知识点

- 基础聚合函数：`COUNT`/`COUNT(DISTINCT)`/`SUM`/`MIN`/`MAX`场景化使用

- `UNION ALL`结果集合并

- 无表关联的基础数据探查方法论

### SQL 代码

```python
import pandas as pd
import sqlite3

# 1. 连接SQLite数据库
conn = sqlite3.connect("/kaggle/input/datasets/tclaim/retail-sales/retail_sales.db")

# 4. 封装通用SQL执行函数，自动打印表格结果
def query_sql(sql):
    return pd.read_sql(sql, conn)
    
print("数据导入完成，调用 query_sql('SQL语句') 即可执行查询")
```

```python
sql = '''
SELECT '总订单数' AS 指标名称, COUNT(订单ID) AS 指标值 FROM "销售订单表"
UNION ALL
SELECT '总商品数' AS 指标名称, COUNT(DISTINCT 商品ID) AS 指标值 FROM "商品信息表"
UNION ALL
SELECT '总客户数' AS 指标名称, COUNT(DISTINCT 客户ID) AS 指标值 FROM "客户信息表"
UNION ALL
SELECT '总门店数' AS 指标名称, COUNT(DISTINCT 门店ID) AS 指标值 FROM "门店信息表"
UNION ALL
SELECT '数据起始日期' AS 指标名称, MIN(销售日期) AS 指标值 FROM "销售订单表"
UNION ALL
SELECT '数据结束日期' AS 指标名称, MAX(销售日期) AS 指标值 FROM "销售订单表";
'''
query_sql(sql)
```

### 结果说明

输出 6 项核心业务指标，清晰展示业务整体规模、数据覆盖周期，是数据分析的第一步标准探查动作，可直接用于业务复盘开篇
