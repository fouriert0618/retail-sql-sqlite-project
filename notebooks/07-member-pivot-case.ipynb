## 模块 7：会员等级月度营收透视表（行转列）

### 业务需求

基于原始表，统计每个月各会员等级的营收情况，实现行转列透视表效果

### 思路分析

1. 直接关联销售订单表、客户信息表，按月份 + 会员等级分组聚合营收

2. 用`CASE WHEN`条件聚合实现行转列，将会员等级转为列

3. 按月份分组，输出每个月各会员等级的营收数据

### 核心知识点

- 条件聚合`CASE WHEN`

- 行转列透视表实现

- 多维度分组聚合

## SQL 代码

```python
import pandas as pd
import sqlite3

# 1. 连接SQLite数据库
conn = sqlite3.connect("/kaggle/input/datasets/tclaim/retail-sales/retail_sales.db")

# 4. 封装通用SQL执行函数，自动打印表格结果
def query_sql(sql):
    return pd.read_sql(sql, conn)
    
print("数据导入完成，调用 query_sql('SQL语句') 即可执行查询")
```

```python
sql = '''
SELECT
    STRFTIME('%Y-%m', o.销售日期) AS 统计月份,
    SUM(CASE WHEN c.会员等级 = '普通会员' THEN o.销售数量 * o."单价(元)" ELSE 0 END) AS 普通会员营收,
    SUM(CASE WHEN c.会员等级 = '银卡会员' THEN o.销售数量 * o."单价(元)" ELSE 0 END) AS 银卡会员营收,
    SUM(CASE WHEN c.会员等级 = '金卡会员' THEN o.销售数量 * o."单价(元)" ELSE 0 END) AS 金卡会员营收,
    SUM(CASE WHEN c.会员等级 = '钻石会员' THEN o.销售数量 * o."单价(元)" ELSE 0 END) AS 钻石会员营收,
    SUM(o.销售数量 * o."单价(元)") AS 月度总营收
FROM "销售订单表" o
INNER JOIN "客户信息表" c ON o.客户ID = c.客户ID
GROUP BY 统计月份
ORDER BY 统计月份;
'''
query_sql(sql)
```

### 结果说明

输出月度会员营收透视表，清晰展示每个月不同会员等级的营收贡献，可直接用于会员运营效果分析、会员等级体系优化
